```python
days=[1, 2, 3, 4, 5]

sleeping=[7, 8, 6, 11, 7]
eating = [2, 3, 4, 3, 2]
working= [7, 8, 7, 2, 2]
playing= [8,5, 7,8,13]
style.use('ggplot')
plt.plot([],[],color='m',label='sleeping')
plt.plot([],[],color='g',label='eating')
plt.plot([],[],color='r',label='working')
plt.plot([],[],color='b',label='playing')
plt.legend()


plt.stackplot(days,sleeping,eating,working,playing,colors=['m','g','r','b'])
plt.xlabel('Hour',color='b')
plt.ylabel('no. of employees',color='b')
plt.title('Stackplot')
plt.grid(color='k')
plt.show()
```

```python
x=[7, 2, 2, 13]
activities=['sleeping','eating','working','playing']
plt.pie(x,labels=activities,colors=['m','c','r','g'],startangle=90,shadow=True,explode=(0.1,0,0,0),autopct=('%1.1f%%'))
plt.legend()
plt.title('Pie Chart')
autopct=()
plt.show()
```

```python
def f(t) :
    return np.exp(-t)*np.cos(2*np.pi*t)
t1=np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.02)

plt.subplot(221)
plt.plot(t1,f(t1),'ro',t2,f(t2))
plt.grid(color='k')
plt.title('Plot 1')
plt.xlabel('x axis',color='b')
plt.ylabel('y axis',color='b')
plt.subplot(222)
plt.plot(t2,f(t2),'bo')
plt.grid(color='k')
plt.title(' Plot 2')
plt.xlabel('x axis',color='b')
# plt.ylabel('y axis',color='b')
plt.show()
```

```python
data=pd.read_csv('countries.csv')
style.use('ggplot')
data_2007=data[data.year==2007]
p=0
for i in data_2007.population:
        p=p+i
        
india=data_2007[data_2007.country=="India"]
china=data_2007[data_2007.country=="China"]
us=data_2007[data_2007.country=="United States"]
c=int(china.population)
i=int(india.population)
o=int(us.population)
x=[c,i,o,p-(c+i+o)]
labelss=['China','India','US','Other country']
plt.pie(x,labels=labelss,colors=['b','r','g','c'],explode=(0,0.1,0,0),shadow=True,autopct=('%1.1f%%'))
plt.title("India vs China vs US vs Rest of the World's population  ",color='b')
plt.show()
```
